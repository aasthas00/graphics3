```python

# Step 1: URL se response lena
url = 'http://books.toscrape.com/catalogue/page-1.html'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Step 2: Sabhi books ke container dhundhna
books = soup.find_all('article', class_='product_pod')

# Step 3: Har book ke liye data extract karna
for book in books:
    # Title (Book name)
    title = book.h3.a['title']

    # Rating (Star rating ko class se extract karte hain)
    rating_class = book.p['class']
    rating = rating_class[1]  # Example: ['star-rating', 'Three'] → 'Three'

    # Price
    price = book.find('p', class_='price_color').text.strip()

    # Print simulated review-style output
    print(f"👤 Customer Name: Anonymous")  # Dummy as site has no customer data
    print(f"⭐ Rating: {rating} Stars")   # Star rating (word format)
    print(f"💬 Review: No actual reviews, just book data used here.")  # Placeholder
    print(f"🏷️ Tag: Price - {price}")  # Tag used as price here
    print("-" * 60)
```
